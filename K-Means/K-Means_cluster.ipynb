### K- Means Clustering 

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sn
import warnings


warnings.filterwarnings('ignore')

income_df = pd.read_csv('Income Data.csv')



sn.lmplot("age","income", data = income_df, fit_reg = False, size = 5)

income_df.head()
```

```python
cluster = KMeans(3)

cluster.fit(income_df)

income_df['labels'] = cluster.labels_

income_df.head()

markers = ['+','^','.']

sn.lmplot( "age", "income", data=income_df, hue = "labels",
            fit_reg=False, markers = markers, size = 5 );
```

```python
scaler = StandardScaler()

income_df_scaled = scaler.fit_transform(income_df)

income_df_scaled[0:5]

cluster_new = KMeans(3, random_state = 42)

cluster_new.fit(income_df_scaled)

income_df['labels_new'] = cluster_new.labels_

markers = ['+','*','^']

sn.lmplot( "age", "income", data=income_df, hue = "labels_new",
            fit_reg=False,markers = markers,size = 4 );

```

### By scaling the data points the seperation the clusters are well segregated.
