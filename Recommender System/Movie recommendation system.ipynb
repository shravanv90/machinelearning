## Item based collaborative filtering

Movie Lens data set 

943 users
1682 items
100000 ratings

```python
import pandas as pd

# creating a movie dataframe
r_cols = ['user_id', 'movie_id', 'rating']
rate_df = pd.read_csv('ml-100k/u.data', sep = '\t', names = r_cols , usecols = range(3))
rate_df.head()

m_cols = ['movie_id', 'title']
mov_df = pd.read_csv('ml-100k/u.item', sep = '|', names = m_cols , usecols = range(2),encoding="ISO-8859-1")

movie_df = pd.merge(rate_df,mov_df)
movie_df.tail()
```

```python
# create a matrix of item relation

user_rate = movie_df.pivot_table(index=movie_df['user_id'],columns = movie_df['title'],values = 'rating')
user_rate.head()
```

```python
# creating a correlation matrix
# Minimum number of observations required per pair of columns to have a valid result is kept to 100.

movie_corr = user_rate.corr(min_periods = 100)

movie_corr.head()
```

```python
myRatings = user_rate.loc[0].dropna()
myRatings
```

```python
sim_rating = pd.Series()

for i in range(len(myRatings.index)):
    sims = movie_corr[myRatings.index[i]].dropna()
    #scaling to give strength to movies which the user has rated higher
    sims = sims.map(lambda x: x * myRatings[i])
    sim_rating=sim_rating.append(sims)

sim_rating.sort_values(inplace = True, ascending = False)
#sim_rating.head()

sim_rating=sim_rating.groupby(sim_rating.index).sum()
sim_rating.sort_values(inplace = True, ascending = False)
sim_rating = sim_rating.drop(myRatings.index)
sim_rating.head(10)
```
